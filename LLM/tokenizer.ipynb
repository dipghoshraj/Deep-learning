```python
from datasets import load_dataset
from transformers import GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token
```

```python
ds = load_dataset("DipamSoni/custom_text_to_sql_dataset", split="train")
```

```python
print(ds[0])
```

```python
def format_and_tokenize(obj):
    return f"{obj['instruction']}\n{obj['input']}\n=> {obj['response']}"
```

```python
def tokenize_function(batch):
    prompts = [
        f"{i}\n{inp}\n=> {resp}"
        for i, inp, resp in zip(batch["instruction"], batch["input"], batch["response"])
    ]
    encoded = tokenizer(
        prompts,
        truncation=True,
        padding="max_length",
        max_length=512,
        return_tensors=None
    )
    encoded["labels"] = encoded["input_ids"]
    return encoded
```

```python
print(tokenize_function(ds[0]))
```

```python
tokenized_dataset = ds.map(tokenize_function, batched=True)
```
